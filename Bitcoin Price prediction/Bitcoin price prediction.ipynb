## Bitcoin price prediction

```python
#importing libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#load dataset
df= pd.read_csv(r'C:\MLCOurse\BitcoinPrice.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.drop(['Date'], 1, inplace= True)
```

```python
df.head(35)
```

```python
#predicting 30 days in future
p_days = 30
df['Prediction']=df[['Price']].shift(-p_days)
```

```python
df.head(30)
```

```python
df.tail()
```

```python
# # seperate X and Y

X= np.array(df.drop(['Prediction'], 1))
X= X[: len(df) - p_days]

```

```python
X.shape
```

```python
X
```

```python
# creating Y as np array
Y = np.array(df['Prediction'])
Y= Y[: -p_days]
```

```python
Y.shape
```

```python
Y
```

```python
# split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
P_days_array = np.array(df.drop(['Prediction'], 1))[-p_days: ]
```

```python
print(P_days_array)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
Rf= RandomForestRegressor(n_estimators = 1000, random_state = 1)
Rf.fit(X_train, y_train)
print('Random Forset Accuracy: {:.2f}%'.format(Rf.score(X_test, y_test)*100))
```

```python
# Prediction
%matplotlib inline
# Rf_predict = Rf.predict(X_test)
# print(Rf_predict)


Rf_predict = Rf.predict(X_test)
plt.figure(figsize=(10,8))
plt.plot(Rf_predict, label='predict')
plt.plot(y_test, label='true')
plt.legend()
plt.show()
```
